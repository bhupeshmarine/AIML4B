## 2. Why Environment Setup Matters

A proper environment setup helps you:

- install the correct Python version
- keep project dependencies organized
- avoid package conflicts
- run notebooks smoothly
- and make your learning reproducible


## 3. Main Tools We Will Use

For this course, the main tools are:

- **Python** for coding
- **Jupyter Notebook** for learning and experimentation
- **VS Code** as an editor (optional but useful)
- **Anaconda** or **pip + venv** for environment management
- **Git and GitHub** later for version control


## 4. What is Jupyter Notebook?

Jupyter Notebook is an interactive environment where you can combine:

- markdown explanations
- Python code
- outputs
- charts
- notes and experiments

This makes it ideal for learning AI and ML.

## 5. Types of Cells in Jupyter

There are mainly two types of cells you will use:

### Markdown Cell
Used for writing explanations, headings, notes, and theory.

### Code Cell
Used for writing and running Python code.


## 6. How to Run a Cell

Common ways to run a cell:

- `Shift + Enter` → run cell and move to next
- `Ctrl + Enter` → run cell and stay on current cell
- `Alt + Enter` → run cell and create a new cell below


## 7. Recommended Setup Options

You can choose one of these:

### Option 1: Anaconda
Good for beginners because many packages are easy to manage.

### Option 2: Python + pip + virtual environment
Lighter and cleaner if you want more control.

### Option 3: Google Colab
Useful if you do not want local installation initially.


## 8. Important Packages for This Course

We will use these core packages:

## 9. Example Installation Commands

If using `pip`, common install commands are:

## 10. Creating a Virtual Environment

A virtual environment helps keep your project packages separate.

Example commands:

```bash

source aiml_env/bin/activate
```

## 11. Checking Python Version

It is often useful to check which Python version is active.

````python
import sys

print('Python version:', sys.version)
````

## 12. Checking Installed Libraries

Run the following to check whether the required libraries are available.

````python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
import sklearn

print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', matplotlib.__version__)
print('Seaborn version:', sns.__version__)
print('Scikit-learn version:', sklearn.__version__)
````

## 15. Common Beginner Problems

Some common problems during setup are:

- Python not installed correctly
- wrong environment active
- package not found errors
- Jupyter using a different kernel
- package version conflicts


## 16. Good Practices

Follow these good habits from the beginning:

- keep one environment per project if needed
- install only required libraries
- name notebooks clearly
- organize files into folders
- save work frequently
- add markdown explanations, not just code
