## 2. What is Artificial Intelligence?

**Artificial Intelligence (AI)** is a broad field focused on building systems that can perform tasks which normally require human intelligence.

Examples:
- voice assistants
- recommendation systems
- fraud detection
- image recognition
- chatbots


## 3. What is Machine Learning?

**Machine Learning (ML)** is a subset of AI where systems learn patterns from data instead of being explicitly programmed for every rule.

Example:
- Instead of writing all rules to detect spam email,
- we train a model on examples of spam and non-spam emails.


## 4. What is Deep Learning?

**Deep Learning** is a subset of Machine Learning that uses neural networks with many layers.

It is commonly used in:
- image classification
- speech recognition
- NLP and large language models
- advanced prediction systems


## 5. AI vs ML vs Deep Learning

| Term | Meaning |
|---|---|
| AI | Broad field of making machines intelligent |
| ML | Machines learn from data |
| Deep Learning | ML using deep neural networks |


## 6. Types of Machine Learning

### Supervised Learning
We have input data and known output labels.

Examples:
- house price prediction
- loan default prediction
- customer churn prediction

### Unsupervised Learning
We have data but no labels.

Examples:
- customer segmentation
- anomaly detection

### Reinforcement Learning
An agent learns by interacting with an environment and receiving rewards or penalties.
