## 6. What is Artificial Intelligence?

**Artificial Intelligence (AI)** is a broad field focused on building systems that can perform tasks which normally require human intelligence.

Examples:
- voice assistants
- recommendation systems
- fraud detection
- image recognition
- chatbots


## 7. What is Machine Learning?

A classic formal definition of Machine Learning is:

> A computer program is said to learn from experience **E** with respect to some class of tasks **T** and performance measure **P**, if its performance at tasks in **T**, as measured by **P**, improves with experience **E**.

### Simple Definition
**Machine Learning (ML)** is a subset of AI where systems learn patterns from data instead of being explicitly programmed for every rule.

### Example
Instead of writing every rule to detect spam emails, we train a model using examples of:
- spam emails
- non-spam emails

Over time, the model learns patterns from the data and improves its ability to classify new emails correctly.

## 8. What is Deep Learning?

**Deep Learning** is a subset of Machine Learning that uses neural networks with many layers.

It is commonly used in:
- image classification
- speech recognition
- NLP and large language models
- advanced prediction systems
