## 7. Virtual Environment

A **virtual environment** is an isolated workspace for a Python project where we can install its own Python packages separately from other projects. It is required because different projects may need different versions of the same library.  
For example, 
**Project A** may need `pandas 2.2`, while **Project B** may need `pandas 1.5`. If both are installed in the same global environment, they can conflict. A virtual environment avoids this problem by keeping each project separate.

When we work on multiple projects, we may need different versions of Python and its libraries. So we create different **virtual environments** where we can install different versions for different projects. That is why we create different environments.




````python
# you can check how many different virtual environments are install on your system with the below command
# you will atleast see one environment like -----------> base                 * /opt/anaconda3
# you may see more depending on how many environments yo have created
!conda env list
````

## 8. What is yaml file
An **`environment.yml`** file is very useful for managing virtual environments because it stores all the important details of the environment in one place.

It can include:
- the environment name
- Python version
- library names
- library versions

### Example of a YAML/YML file

```yaml
name: aiml_env
channels:
  - conda-forge
dependencies:
  - python=2.9
  - numpy=2.0
  - pandas=2.2
```

In this project, in our current working directory, you can go and check that we already have an **`environment.yml`** file. Open that file and see what information is stored inside it.

### Why we need yaml file
To manage any project easily, we use an **`environment.yml`** file. These files are used to create an environment with one command. It will install Python with a specific version and the required libraries with one line of command.


So instead of installing everything manually one by one, we can write all required versions in one file and create the environment from it.



This makes it easy to recreate the same environment again without installing everything manually one by one.
