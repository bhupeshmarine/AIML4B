# Data Types in Python

In Python, data can be stored in different forms. These forms are called **data types**.

Different data types are needed because different kinds of data behave differently.  
For example:

- a **name** is text
- an **age** is a number
- a **price** can be a decimal value
- a **count** of someting is integar
- multiple values can be stored in a collection

Using different data types helps Python understand:
- what kind of data is being stored
- what operations can be performed on that data

## Some Important Python Data Types

- **int** → whole numbers like `10`, `25`
- **float** → decimal numbers like `10.5`, `99.99`
- **str** → text like `"Bhupesh"`
- **bool** → `True` or `False`
- **list** → ordered collection of items: `['hello', 1, 'Bhupesh', 31]`
- **dict** → data stored in key-value pairs: `{
    "name": "Bhupesh",
    "roll": 1234,
    "city": "Dallas"
}`

# List in Python

A list stores multiple items in one variable.

Lists are written using square brackets `[]`.

```python
fruits = ["apple", "banana", "mango"]

print(fruits)
```

## List can store different data types

A list can contain strings, numbers, and other values.

```python
data = ["Bhupesh", 1234, "Dallas", 95.5]

print(data)
```

## Access items from a list

Each item in a list has an index.

Index starts from `0`.

```python
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

## Add an item to a list

Use `append()` to add a new item at the end.

```python
fruits.append("orange")

print(fruits)
```

## Change an item in a list

We can update a value using its index.

```python
fruits[1] = "grapes"

print(fruits)
```

## Remove an item from a list

Use `remove()` to delete an item.

```python
fruits.remove("apple")

print(fruits)
```

## Length of a list

Use `len()` to find how many items are in the list.

```python
print(len(fruits))
```

## Create a DataFrame from a list

A list can be converted into a pandas DataFrame.

```python
import pandas as pd

names = ["Bhupesh", "Rahul", "Aman"]

df = pd.DataFrame(names, columns=["Name"])
print(df)
```

## List of lists

A list can also contain other lists.

This is useful for table-like data.

```python
student_data = [
    ["Bhupesh", 101, "Dallas"],
    ["Rahul", 102, "Boston"],
    ["Aman", 103, "Austin"]
]

print(student_data)
```

## Create a DataFrame from a list of lists

Each inner list becomes one row in the DataFrame.

```python
df2 = pd.DataFrame(student_data, columns=["Name", "Roll", "City"])
df2
```

# Dictionary in Python

A dictionary stores data in **key-value pairs**. They are written using `{}

- **Key** = label
- **Value** = data

Example:
- `"name"` is a key
- `"Bhupesh"` is a value

```python
student = {
    "name": "Bhupesh",
    "roll": 1234,
    "city": "Dallas"
}

print(student)
```

## Access values from a dictionary

We use the key to get its value.

```python
print(student["name"])
print(student["roll"])
print(student["city"])
```

## Dictionary with lists

In a dictionary, one value can also be a list.

This is useful when we want to store multiple rows of data under each key.

```python
student_data = {
    "name": ["Bhupesh", "Rahul", "Aman"],  # key and list of values for each key
    
    "roll": [101, 102, 103],
    
    "city": ["Dallas", "Boston", "Austin"]
}

print(student_data)
```

```python
student_data['name']
```

```python
#acessing first value of the name
student_data['name'][0]
```

## Create a DataFrame from a dictionary with lists

When dictionary values are lists:
- keys become column names
- list items become rows

```python
df = pd.DataFrame(student_data)
df
```
